```
#import library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```
#load csv and show most comman 5 row
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```
#show info from dataframe, ex show column null
df.info()
```

```
#drop row user id
data = df.drop(columns=['User ID'])

#implement one hot encoding
data = pd.get_dummies(data)
data.head()
```

```
#separate attribute and label
predictions = ['Age', 'EstimatedSalary', 'Purchased','Gender_Female', 'Gender_Male']
x = data[predictions]
y = data['Purchased']
```

```
#split train set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```
from sklearn import linear_model
#make model with Logistic Regression
model = linear_model.LogisticRegression()

#training model
model.fit(x_train, y_train)
```

```
#show score with test set
model.score(x_test, y_test)
```
