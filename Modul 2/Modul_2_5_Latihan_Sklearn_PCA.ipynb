```
from sklearn import datasets
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
atribut = iris.data
label = iris.target

#split data
X_train, X_test, y_train, y_test = train_test_split(atribut, label, test_size=0.2)
```

##Tanpa PCA

```
from sklearn import tree

decision_tree = tree.DecisionTreeClassifier()
model_pertama = decision_tree.fit(X_train, y_train)
model_pertama.score(X_test, y_test)
```

##Menggunakan PCA

```
from sklearn.decomposition import PCA

#Create object PCA with 4 principal component
pca = PCA(n_components=4)

#applied pca in dataset
pca_atributes = pca.fit_transform(X_train)

#Show variance in atribute
pca.explained_variance_ratio_
```

```
#Get pca with 2 principal component
pca = PCA(n_components=2)

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

```
model2 = decision_tree.fit(X_train_pca, y_train)
model2.score(X_test_pca, y_test)
```
