```
#import library
import sklearn
from sklearn import datasets
from sklearn.model_selection import cross_val_score
from sklearn import tree
```

```
#load iris dataset
iris = datasets.load_iris()

#separate label and attribute from dataset
x = iris.data
y = iris.target
```

```
#make model using decission tree
clf = tree.DecisionTreeClassifier()
```

```
"""evaluation performance model using cross validation 
clf is ML model, x is attribut, y is label, cv is number of folds
"""
scores = cross_val_score(clf, x, y, cv=5)
```

```
"""result of validation show in below,
range value score indicates that our model is performing well
"""
scores
```
